```python
import pandas as pd
from pandas import Series
import numpy as np
import tabula
import bamboolib as bam
from datetime import date
```

```python
fec_election_info = pd.read_excel('fec_election_info.xlsx')
```

```python
tmp_condition = fec_election_info['SENATE RACE'].isin(['S'])
fec_election_info.loc[tmp_condition, 'SENATE RACE'] = 'YES'
fec_election_info.loc[~tmp_condition, 'SENATE RACE'] = 'NO'
for column_name in ['STATE', 'SENATE RACE']:
    fec_election_info[column_name] = fec_election_info[column_name].astype('string')
fec_election_info['CONGRESSIONAL PRIMARY DATE'] = pd.to_datetime(fec_election_info['CONGRESSIONAL PRIMARY DATE'], format='%Y-%m-%d')
fec_election_info.columns = fec_election_info.columns.str.replace(' ', '_')
fec_election_info = fec_election_info.sort_values(by=['CONGRESSIONAL_PRIMARY_DATE', 'STATE'], ascending=[True, True])
fec_election_info.to_excel('federal_election_info.xlsx')
```
